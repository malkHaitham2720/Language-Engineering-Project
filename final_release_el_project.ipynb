```python
import pandas as pd
```

```python
import nltk
nltk.download('punkt') #can split text into sentences and words.
nltk.download('averaged_perceptron_tagger') # assign grammatical tags to words, such as "noun" or "verb".
nltk.download('wordnet') #a lexical database for English 
nltk.download('stopwords') #  remov sotp_words  from text during NLP preprocessing. (such as "the", "and", "a", etc.) 
```

```python
from nltk.stem import WordNetLemmatizer#Lemmatization is the process of reducing a word to its base or dictionary form
from nltk.corpus import wordnet# lexical database for English. is a large semantic network of words and their relationships, such as synonyms and antonyms, which can be used for tasks such as word sense disambiguation.
from nltk.corpus import stopwords#
stop_words=stopwords.words("english")
lemmatizer = WordNetLemmatizer()
```

loading data


```python
df=pd.read_csv('dataset2.csv',encoding='utf-8')
df.head(10)
```

```python
x = df["English"][:5000]# input 
y = df["Arbic"][:5000]# output
```

```python
def nltk_tag_to_wordnet_tag(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:          
        return None
```

```python
def lemmatize_sentence(sentence):
    nltk_tagged = nltk.pos_tag(nltk.word_tokenize(sentence))  
    wordnet_tagged = map(lambda x: (x[0], nltk_tag_to_wordnet_tag(x[1])), nltk_tagged)
    lemmatized_sentence = []
    for word, tag in wordnet_tagged:
      if word not in stop_words:
        if tag is None:
            lemmatized_sentence.append(word)
        else:        
            lemmatized_sentence.append(lemmatizer.lemmatize(word, tag))
    return " ".join(lemmatized_sentence)
```

Dealing with missing values in cells

```python
# Count the number of missing values in each column
print(df.isnull().sum())
# Fill missing values with a specified value
df.fillna(value=0, inplace=True)
# Drop rows that contain missing values
df.dropna(inplace=True)
print(df.head(5))
```

```python
x_cleaned = x.apply(lemmatize_sentence)
```

ML

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(encoding='latin-1', ngram_range=(1, 2), stop_words='english')
tfidf = tfidf.fit(x_cleaned)
X_tfidf = tfidf.transform(x_cleaned)
X_tfidf.shape
```

spliting our data for training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size= 0.2)
```
